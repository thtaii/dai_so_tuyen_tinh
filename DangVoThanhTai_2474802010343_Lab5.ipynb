```python
#2. Ứng dụng 2 –Liên phân số và số 
import numpy as np

c = [3, 7, 15, 1, 292, 1, 1, 1, 2, 1, 3, 1, 14, 2]  # dãy phân số biểu diễn Pi

M = np.asmatrix([[1, 2], [3, 4]])  # khởi tạo ma trận 2x2 (không dùng ngay)

for i in range(len(c)):
    ci = np.asmatrix([[1, 1], [1, 0]])
    ci[0, 0] = c[i]
    if i == 0:
        M = ci
    else:
        M = M.dot(ci)

print("Ma trận M sau khi nhân liên tục:")
print(M)

pi_approx = M[0, 0] / M[1, 0]
print("Giá trị xấp xỉ số Pi tính được:", pi_approx)
```

```python
#3. Về phân tích hồi quy tuyến tính 
import numpy as np

# Nhập dữ liệu X (ma trận 5x2), mỗi hàng là một điểm dữ liệu, cột đầu là 1
X = np.array([
    [1, 1],
    [1, 2],
    [1, 3],
    [1, 4],
    [1, 5]
])

# Vector kết quả Y
Y = np.array([1, 2, 4, 4, 6])

# Tính chuyển vị của X
XT = X.T

# Tính A1 = (X^T X)^(-1)
A1 = np.linalg.inv(XT.dot(X))

# Tính A2 = X^T Y
A2 = XT.dot(Y)

# Tính vector A = A1.dot(A2)
A = A1.dot(A2)

print("Vector nghiệm A (chứa intercept và hệ số góc):")
print(A)

# Phương trình hồi quy gần đúng là:
# Y = A[0] + A[1]*X

# Tính giá trị dự đoán Y_pred
Y_pred = X.dot(A)

# Tính sai số là tổng bình phương sai số
error = np.sum((Y - Y_pred) ** 2)
print("Tổng bình phương sai số:", error)
```

```python
import numpy as np

def bpcuctieu(a):
    b = []
    o1 = []
    for i in range(len(a)):
        o1.append(1.0)        # cột hệ số chệch (intercept)
        b.append(i * 1.0)     # cột biến độc lập (thời gian)

    # Xây dựng ma trận dữ liệu (2 x n) gồm cột 1 và cột biến thời gian
    x = np.array([o1, b])    # kích thước (2, n)

    # Tính u = x . x^T
    u = np.dot(x, x.T)       # kích thước (2, 2)

    # Vector kết quả y (lượng thuốc)
    y = np.array(a)          # kích thước (n,)

    from numpy import linalg

    # Tính nghịch đảo của u
    u_1 = np.linalg.inv(u)

    # Tính v = u_1 . x
    v = np.dot(u_1, x)       # kích thước (2, n)

    # Tính hệ số A = v . y
    A = np.dot(v, y)

    return A

# Dữ liệu lượng thuốc theo thời gian
dulieu = [10.0, 8, 7, 5, 2]

# Gọi hàm
A = bpcuctieu(dulieu)

print(f"Hệ số phương trình tuyến tính y = a0 + a1*x là:")
print(f"a0 (intercept) = {A[0]}")
print(f"a1 (hệ số góc) = {A[1]}")

print(f"Phương trình gần đúng: y = {A[0]:.2f} + ({A[1]:.2f}) * x")
```
